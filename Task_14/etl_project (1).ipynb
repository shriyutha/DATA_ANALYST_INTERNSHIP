## STEP 3: TRANSFORM

### Standardize Column Names:


```python
data.columns = (data.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('-', '_'))

data.columns
```

### Remove Duplicates:

```python
before = len(data)

data = data.drop_duplicates()

after = len(data)

print('Duplicates Removed:', before - after)
```

```python
data.info()
```

### Fix Data Types:

```python
data['date'] = pd.to_datetime(data['date'])
```

```python
data['amount'] = pd.to_numeric(data['amount'])
data['qty'] = pd.to_numeric(data['qty'])
```

### Create Derived Business Columns:

```python
data['order_type'] = data['b2b'].apply(lambda x: 'B2B' if x == True else 'B2C')
```

```python
data['revenue'] = data['amount'] * data['qty']
```

```python
data['order_month'] = data['date'].dt.to_period('M')
```

### Handle Missing Values:

```python
data = data.drop(columns = ['index', 'unnamed:_22', 'sales_channel'], errors = 'ignore')
```

```python
data = data.dropna(subset = ['order_id', 'date', 'amount', 'qty', 'ship_city', 'ship_state', 'ship_postal_code', 'ship_country'])
```

```python
data = data.fillna('Unknown')
```

## STEP 4: Load - Output
## Save Cleaned Data to CSV

```python
data.to_csv('data/processed/cleaned_amazon_store.csv', index = False)
```

## Create Dimension Tables

### Order Table:

```python
orders = data[['order_id', 'date', 'status', 'fulfilment', 'order_type', 'asin', 'revenue']]
```

### Products Table:

```python
products = data[['sku', 'style', 'category', 'size', 'asin']]
```

### Shipping Table:

```python
shipping = data[['order_id', 'ship_city', 'ship_state', 'ship_country', 'ship_service_level', 'courier_status']]
```

## STEP 5: VALIDATION & Referential Integrity Check
Check Row Counts

```python
print('Shipping:', len(shipping))
print('Orders:', len(orders))
print('Products:', len(products))
```

```python
assert shipping['order_id'].isin(orders['order_id']).all()
print('Foreign key validation.')
```

## Load to PostgreSQL

```python
from sqlalchemy import create_engine
from urllib.parse import quote_plus


username = 'postgres'
password = ' '
host = 'localhost'
port = '5432'
database = 'etl_project'


password_encoded = quote_plus(password)
engine = create_engine(f'postgresql://{username}:{password_encoded}@{host}:{port}/{database}')
```

```python
orders.to_sql('amazon_orders', con = engine, if_exists = 'replace', index = False)
products.to_sql('amazon_products', con = engine, if_exists = 'replace', index = False)
shipping.to_sql('amazon_shipping', con = engine, if_exists = 'replace', index = False)

print('Data loaded into PostgreSQL')
```
