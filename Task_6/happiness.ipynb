```python
# Pie Chart: Top 10 Countries by Happiness Score:

df = data.sort_values('Score', ascending = False)[:10]

plt.figure(figsize = (10, 5))
plt.pie(df['Score'], autopct = '%1.1f%%', startangle = 120, labels = df['Country or region'])
plt.title('Top 10 Countries Share of Happiness Score')
plt.show()
```

```python
# Line Chart: Happiness Score Trend by Rank:

df = data.sort_values('Overall rank')

plt.figure(figsize = (30, 15))
plt.plot(df['Overall rank'], df['Score'], marker = '*', color = 'black', )
plt.title('Happiness Score Trend by Country Rank')
plt.xlabel('Overall Rank')
plt.ylabel('Happiness Score')
plt.grid(True)
plt.show()
```

```python
# Histogram: Distribution of Happiness Scores:


plt.figure(figsize = (10, 5))

plt.hist(df['Score'], bins = 15, color = 'black', label = 'Number of Countries')

plt.hist(df['Score'], bins = 15, color = 'red', histtype = 'step', density = True, linewidth = 2.5, label = 'Distribution Shape')
plt.title('Distribution of Happiness Scores')
plt.xlabel('Happiness Score')
plt.ylabel('Number of Countries')
plt.grid(True)
plt.legend()
plt.show()


```

```python
# Scatter Plot: GDP per Capita vs Happiness Score:

plt.figure(figsize = (10, 5))

plt.scatter(data['GDP per capita'], data['Score'], color = 'black', marker = '.', label = 'Countries')

x = data['GDP per capita']
y = data['Score']
m, b = np.polyfit(x, y, 1)   # slope and intercept
plt.plot(x, m*x + b, linestyle = '--', label = 'Regression Line', color = 'red')

plt.title('GDP per Capita vs Happiness Score (2019)')
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.grid(True)
plt.show()
```
